# Ciroyu(turnover)gözden geçirme.

```python
front = df['turnover']
df.drop(labels=['turnover'], axis=1,inplace = True)
df.insert(0, 'turnover', front)
df.head()
```

Bu kod,bir pandas DataFrame'de('df')yer alan 'turnover' sütununa göre gruplama yaparak,herbir 'turnover'değerinin veri setindeki oranını hesaplar.

```python
turnover_rate = df.groupby('turnover').size() / len(df)
turnover_rate
```
